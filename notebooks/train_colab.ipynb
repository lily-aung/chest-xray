```python
import pandas as pd

TRAIN_CSV_IN = "/content/drive/MyDrive/ChestXRay/train.csv"
VAL_CSV_IN   = "/content/drive/MyDrive/ChestXRay/val.csv"
TEST_CSV_IN  = "/content/drive/MyDrive/ChestXRay/test.csv"

TRAIN_CSV_OUT = "/content/drive/MyDrive/ChestXRay/train_colab.csv"
VAL_CSV_OUT   = "/content/drive/MyDrive/ChestXRay/val_colab.csv"
TEST_CSV_OUT  = "/content/drive/MyDrive/ChestXRay/test_colab.csv"

OLD_PREFIX = "/Users/lily/Documents/Projects/chest-xray/data/raw/archive/"
NEW_PREFIX = "/content/chestxray/data"

def remap_csv(in_csv, out_csv):
    df = pd.read_csv(in_csv)

    # Ensure image_path is string
    df["image_path"] = df["image_path"].astype(str)

    # Normalize slashes (important if CSV came from Windows/macOS mix)
    df["image_path"] = df["image_path"].str.replace("\\", "/", regex=False)

    # Replace prefix
    df["image_path"] = df["image_path"].str.replace(
        OLD_PREFIX.rstrip("/") + "/",
        NEW_PREFIX.rstrip("/") + "/",
        regex=False
    )

    df.to_csv(out_csv, index=False)
    print(f"Wrote: {out_csv}")
    print(df[["image_path"]].head(3))

# Run
remap_csv(TRAIN_CSV_IN, TRAIN_CSV_OUT)
remap_csv(VAL_CSV_IN, VAL_CSV_OUT)
remap_csv(TEST_CSV_IN, TEST_CSV_OUT)


CKPT_DIR = "/content/drive/MyDrive/ChestXRay/checkpoints"
os.makedirs(CKPT_DIR, exist_ok=True)

```

```python
import random, numpy as np, pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms
from PIL import Image
import timm
from tqdm import tqdm

def set_seed(seed: int):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

```

```python
import os, re
from pathlib import Path

BASE_CFG = "configs/swin_tiny_patch4_window7_224.yaml"
OUT_CFG  = "configs/swin_tiny_colab_finetune.yaml"

train_csv = "/content/drive/MyDrive/ChestXRay/train_colab.csv"
val_csv   = "/content/drive/MyDrive/ChestXRay/val_colab.csv"
test_csv  = "/content/drive/MyDrive/ChestXRay/test_colab.csv"  # change if needed / or comment out
img_dir   = "null"

text = Path(BASE_CFG).read_text()

def replace_yaml_key(block_text, key, new_value):
    # replaces lines like "  key: something"
    pattern = rf"(^\s*{re.escape(key)}\s*:\s*).*$"
    repl = rf"\1{new_value}"
    return re.sub(pattern, repl, block_text, flags=re.MULTILINE)

# update the data paths
text = replace_yaml_key(text, "train_csv", train_csv)
text = replace_yaml_key(text, "val_csv", val_csv)
text = replace_yaml_key(text, "test_csv", test_csv)
text = replace_yaml_key(text, "img_dir", img_dir)

# ensure finetuning
text = replace_yaml_key(text, "linear_probe", "false")

Path(OUT_CFG).write_text(text)
print("Wrote:", OUT_CFG)

# show the edited data/model sections for confirmation
print("\n--- data section ---")
m = re.search(r"(?ms)^data:\n(.*?)(^\S|\Z)", text)
print(m.group(0) if m else "data section not found")

print("\n--- model section ---")
m = re.search(r"(?ms)^model:\n(.*?)(^\S|\Z)", text)
print(m.group(0) if m else "model section not found")
```
